# Build the final model and generate the predictions for submission 

Exploratory data analysis is already done and several models have been constructed and tested beforehand. In this notebook we will pick a model, and use it to predict on the project test set. Predictions will be finally exported in file for submission.

## Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import xgboost as xgb
```

Set the rng seed.

```python
seed = 17
```

## Import data

```python
df_train = pd.read_csv('../data/train_prepd.csv')
df_test = pd.read_csv('../data/test_prepd.csv')

print('Training set:')
display(df_train.head(5))

print('Testing set:')
display(df_test.head(5))
```

## Create dependent and independent variables 

```python
X_train = df_train.iloc[:, 1:]
y_train = df_train.iloc[:, 0]
X_test = df_test
```

```python
print('X_train: ')
display(X_train.head(5))
print('dims: ', X_train.shape)
print()

print('y_train: ')
display(y_train.head(5))
print('dims: ', y_train.shape)
print()

print('X_test: ')
display(X_test.head(5))
print('dims: ', X_test.shape)
```

## Build and fit the model 

For model details, please see the corresponding notebook inside this repo.

```python
clf = xgb.XGBClassifier(colsample_bytree=0.5,
                        gamma=1,
                        learning_rate=0.3,
                        max_depth=8,
                        min_child_weight=3,
                        reg_lambda=1,
                        scale_pos_weight=1,
                        use_label_encoder=False,
                        random_state=seed
                        )
clf.fit(X_train, y_train)
```

## Predict

```python
y_pred = clf.predict(X_test)
```

## Export to .csv 

```python
# Use the sample file provided by Kaggle and just change the predictions with ours. Force to integer.
submission = pd.read_csv('gender_submission.csv')
submission['Survived'] = y_pred.astype(int)
```

```python
submission.to_csv('submission_xgb.csv', index=False)
```
